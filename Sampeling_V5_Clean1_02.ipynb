```
#extract words from string without punctuation (FINAL) *
def book_punc(i):
  books[i] = re.sub('['+string.punctuation+']', '', str(books[i])).split() 
  #print (books[i]) #check Remove
  #print (len(books[i])) #check Remove
```

```
#removing stop words *
def book_stopwords(i):
  #print(books[i])
  tokenized_words = books[i] 
  #print(tokenized_words)
  stop_words = stopwords.words('english')
  books[i]=[word for word in tokenized_words if word not in stop_words]
  #print (len(books[i]))
  #print (books[i])
```

```
#Stemmer *
def book_stemmer(i):
  ps = PorterStemmer()
  example_words = books[i]
  new = []
  for w in example_words:
   new.append(ps.stem(w))
  books[i]=new
  #print (books[i])
  #print(len(books[i]))
  #print("hey")
```
